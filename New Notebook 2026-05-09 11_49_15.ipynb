```python
af_airports = spark.read.csv(
    "/Volumes/practical/airportdata/airdata/airports.csv",
    header=True,
    inferSchema=True
)
```

```python
# Cast columns
af_airports = af_airports \
    .withColumn("latitude_deg", col("latitude_deg").cast("double")) \
    .withColumn("longitude_deg", col("longitude_deg").cast("double")) \
    .withColumn("elevation_ft", col("elevation_ft").cast("integer"))

# Show schema
af_airports.printSchema()
```

```python
df_runways = spark.read.csv(
    "/Volumes/practical/airportdata/airdata/runways.csv",
    header=True,
    inferSchema=True
)
```

```python
df_runways = df_runways \
    .withColumn("length_ft", col("length_ft").cast("integer")) \
    .withColumn("width_ft", col("width_ft").cast("integer"))

# Show schema
df_runways.printSchema()
```

```python
af_airports.printSchema()
```

```python
print("Airports Row Count:", af_airports.count())
```

```python
df_runways.printSchema()
```

```python
print("Runways Row Count:", df_runways.count())
```

```python
# Inner join between airports and runways dataframes

df_combined = af_airports.join(
    df_runways,
    af_airports.ident == df_runways.airport_ident,
    "inner"
)

```

```python
df_combined.show(5)
```

```python
df_combined.printSchema()
```

```python
print("Combined Row Count:", df_combined.count())
```

```python
df_combined = df_combined.withColumn(
    "runway_category",
    when(col("length_ft") >= 10000, "Long")
    .when((col("length_ft") >= 5000) & (col("length_ft") < 10000), "Medium")
    .otherwise("Short")
)
```

```python
# Show results
df_combined.select(
    "airport_ident",
    "length_ft",
    "runway_category"
).show(10)
```

```python
top_countries = df_combined.groupBy("iso_country") \
    .agg(count("airport_ident").alias("runway_count")) \
    .orderBy(col("runway_count").desc()) \
    .limit(10)
```

```python
# Display result
top_countries.show()
```

```python
avg_runway_length = df_combined.filter(
    col("length_ft").isNotNull()
).groupBy("continent") \
 .agg(avg("length_ft").alias("avg_runway_length_ft")) \
 .orderBy(col("avg_runway_length_ft").desc())
```

```python
# Display result
avg_runway_length.show()
```

```python
# Top 5 airports with longest maximum runway length
top_airports = df_combined.groupBy(
    "name",
    "iso_country"
).agg(
    max("length_ft").alias("max_runway_length_ft")
).orderBy(
    col("max_runway_length_ft").desc()
).limit(5)
```

```python
# Display result
top_airports.show(truncate=False)
```

```python
df_combined.write \
    .mode("overwrite") \
    .parquet("/Volumes/practical/airportdata/airdata/df_combined_parquet")
```
